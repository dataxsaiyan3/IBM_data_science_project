```python
# Importing json to transofrm the data into a json format
import json

# To show the time while iterating 
from tqdm import tqdm

# seaborn and matplotlib were imported for data visualisation
import matplotlib.pyplot as plt
import seaborn as sns

# highlighter is imported to be used as a syntax highlighter 
from pygments import highlight

# Importing JsonLexer to split a JSON String into a list of annotated tokens
from pygments.lexers import JsonLexer

# user for terminal output with ANSI sequences
from pygments.formatters import TerminalFormatter

# Sort reviews and app will be used to get the data from google play store
from google_play_scraper import Sort, reviews, app

%matplotlib inline
%config InlineBackend.figure_format='retina'

# Setting the style of charts
sns.set(style='whitegrid', font_scale=1.2)
```

```python
## We start by looping through
for appName, appId in zip(BrawlStar_clashOfClans['appName'], BrawlStar_clashOfClans['appId']):
    
    # localizing the time and formating the date
    startTime = dt.datetime.now(tz=get_localzone())
    dateFormat = "%m-%d-%y - %T %p"    
    
    # Print starting output for app
    print('#'*60)   
    print(f'* App Name {appName} started at {startTime.strftime(dateFormat)}')
    print()
    
    # Creating an empty list to store reviews
    # apps_rvws here means apps review
    appsReviews = []
    
    # Number of reviews to scrape per batch
    batchSize = 200
    
    # Batch tracking numbers, will be set to 0 first to show the completed batches counts
    batchNum = 0
    
    
    # Here we get reviews and continuation token, this is a google play scraper API
    # rvws here means reviews
    rvws, token = reviews(
        appId,           # found in app's url
        lang='en',        # defaults to 'en'
        country='us',     # defaults to 'us'
        sort=Sort.NEWEST, # start with most recent
        count=batchSize       # batch size
    )
    
    
    # Optaining reviews for each app and adding keys
    # rv stand for one review
    for rv in rvws:
        rv['appName'] = appName
        rv['appId'] = appId
     
    
    # Adding the app reviews to the empity list we created
    appsReviews.extend(rvws)
    
    # Adding 1 to the batch number to track the count of batches
    batchNum +=1 
    # Printing the number of batches
    print(f'Batch number {batchNum} was completed')
    
    # Creating a wait time of 1 to 5 secinds between each batch
    time.sleep(random.randint(1,5))
    
    
    
    # Creating an empty list to append review IDs before starting a new batch
    idsReviews = []
    for review in appsReviews:
        
        idsReviews.append(review['reviewId'])
    
    
    # Here we iterate over the number of batches we need to get as much data as we want
    for batch in range(12000):
        rvws, token = reviews( # store continuation_token
            appId,
            lang='en',
            country='us',
            sort=Sort.NEWEST,
            count=batchSize,
            # Using the token from the previous batch
            continuation_token=token
        )
        
        # Creating an empty for ids reviews number 2 list to append the unique values of reviews IDs
        idsReviews2 = []
        for rv in rvws:
            idsReviews2.append(rv['reviewId'])
            
            # And add keys for name and id to ea review dict
            rv['appName'] = appName # add key for app's name
            rv['appId'] = appId     # add key for app's id
     
        # Add the list of review dicts to main app_reviews list
        appsReviews.extend(rvws)
        
        # Batch tracking numbers, will be set to 0 first to show the completed batches counts
        batchNum +=1
        
        # Here we combine the previews reviews IDs with the new one
        idsReviewsAll = idsReviews + idsReviews2
        if len(set(idsReviews)) == len(set(idsReviewsAll)):
            print(f'Task completed, no more reviews to scrape, stoped at batch number {batchNum}.\n')
            break
        
        # The all reviews will be checked by the rv_ids in order to be able to execute the next batch
        idsReviews = idsReviewsAll
        
        
        # Set at every 100th batch
        if batchNum%100 == 0:
            
            # Show an update of the number of reviews inserted using current time
            print(f'Completed Batch number {batchNum}.')
            
            # Here we insert the data in the MongoDB under the reviews data frame
            appsData.insert_many(appsReviews)
            
            # Show an update of the number of reviews inserted using current time
            current_time = dt.datetime.now(tz=get_localzone())
            print(f"""
            App data inserted {len(appsReviews)} {appName} 
            into MongoDB at {current_time.strftime(dateFormat)}.\n
            """)
            
            # empty our list for next round of 100 batches
            appsReviews = []
        
        # Wait 1 to 5 seconds to start next batch
        time.sleep(random.randint(1,5))
      
    
    # Print update when max number of batches has been reached
      # OR when last batch didn't add any unique reviews
    print(f'Completed scraping {appName}.')
    print(f'There are {len(set(idsReviews))} unique reviews.\n')
    
    
    # Here we again insert the data in the MongoDB under the reviews dataframe
    appsData.insert_many(appsReviews)
    
    # Get end time
    endTime = dt.datetime.now(tz=get_localzone())
    
    # Print ending output for app
    print(f"""
    All data of {appName} reviews has been inserted into MongoDB
    at {endTime.strftime(dateFormat)}.\n
    """)
    print(f'Time elapsed for {appName}: {endTime-startTime}')
    print('---'*20)
    print('---'*20)
    print('\n')
    
    # Wait 1 to 5 seconds to start scraping next app
    time.sleep(random.randint(1,5))
```
